```python
# Robust cell: auto-detect header, map columns (handles typos), aggregate & plot
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from difflib import get_close_matches

# If df already in memory, skip file load. Otherwise load from uploaded path.
uploaded_path = "/mnt/data/data-export (1).csv"

try:
    df  # check if df exists
    print("Using already-loaded df in memory.")
except NameError:
    if not os.path.exists(uploaded_path):
        raise FileNotFoundError(f"CSV not found at {uploaded_path!r}. Please confirm path/upload.")
    # Try to auto-detect header row (handles GA4 metadata above header)
    raw = pd.read_csv(uploaded_path, header=None, dtype=str, keep_default_na=False)
    header_idx = None
    for i in range(min(40, len(raw))):
        row_text = " ".join(raw.iloc[i].str.lower().tolist())
        if any(k in row_text for k in ["channel", "session", "engag"]):
            header_idx = i
            break
    if header_idx is None:
        header_idx = 0
    df = pd.read_csv(uploaded_path, header=header_idx)
    # strip whitespace from column names
    df.columns = df.columns.astype(str).str.strip()

print("Detected columns:")
print(df.columns.tolist())

# --- fuzzy map the needed columns -----------------------------------------
cols = list(df.columns)
cols_lower = [c.lower().strip() for c in cols]

def find_best(names_candidates):
    """Return the actual column name from df.columns that best matches any candidate substrings."""
    for cand in names_candidates:
        # try direct substring match first
        for i,c in enumerate(cols_lower):
            if cand in c:
                return cols[i]
    # fallback: difflib close match
    matches = get_close_matches(names_candidates[0], cols, n=1, cutoff=0.6) if names_candidates else []
    return matches[0] if matches else None

# candidate lists (expand if you know other possible spellings)
channel_candidates = ["channel group", "channelgroup", "channel"]
sessions_candidates = ["sessions", "total sessions", "session"]
engaged_candidates = ["engaged sessions", "enagaged sessions", "engag sessions", "engaged"]

channel_col = find_best(channel_candidates)
sessions_col = find_best(sessions_candidates)
engaged_col = find_best(engaged_candidates)

print("\nMapped columns:")
print(" channel_col ->", channel_col)
print(" sessions_col ->", sessions_col)
print(" engaged_col ->", engaged_col)

if channel_col is None or sessions_col is None or engaged_col is None:
    print("\nCould not auto-map all required columns. Please paste the output of df.columns.tolist() here.")
    raise SystemExit("Missing required columns mapping.")

# --- sanitize numeric columns ------------------------------------------------
def to_int_safe(series):
    s = series.astype(str).str.replace(',', '').str.strip()
    return pd.to_numeric(s, errors='coerce').fillna(0).astype(int)

df[sessions_col] = to_int_safe(df[sessions_col])
df[engaged_col] = to_int_safe(df[engaged_col])

# --- aggregation -------------------------------------------------------------
session_df = df.groupby(channel_col)[[sessions_col, engaged_col]].sum().reset_index()
session_df["Non-Engaged"] = session_df[sessions_col] - session_df[engaged_col]

# rename to consistent names for plotting
session_df = session_df.rename(columns={
    channel_col: "channel group",
    sessions_col: "Sessions",
    engaged_col: "Engaged Sessions"
})

print("\nAggregated table (first 50 rows):")
print(session_df.head(50).to_string(index=False))

# --- prepare melted df for seaborn ------------------------------------------
melted = session_df.melt(id_vars="channel group", value_vars=["Engaged Sessions", "Non-Engaged"],
                         var_name="variable", value_name="value")

# optional: sort channels by total sessions
order = session_df.sort_values("Sessions", ascending=False)["channel group"].tolist()

# --- plot -------------------------------------------------------------------
plt.figure(figsize=(10,6))
sns.barplot(data=melted, x="channel group", y="value", hue="variable", order=order)
plt.title("Engaged vs Non-Engaged Sessions by Channel Group")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of Sessions")
plt.xlabel("Channel Group")
plt.tight_layout()
plt.show()
```
